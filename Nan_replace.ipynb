```python
import pandas as pd
from pycaret.regression import *
```

```python
train = train.drop(['num_date_time'],axis=1)
```

```python
regression_setup = setup(data=train, target='전력소비량(kWh)')
```

```python
regression_model = create_model('xgboost', tree_method='gpu_hist')
```

```python
columns_to_impute = ['강수량(mm)', '풍속(m/s)', '습도(%)', '일조(hr)', '일사(MJ/m2)']
```

```python
rows_with_missing_values = train[train[columns_to_impute].isnull().any(axis=1)]
```

```python
features_for_prediction = rows_with_missing_values.drop(columns=['전력소비량(kWh)'])
```

```python
predicted_values = predict_model(regression_model, data=rows_with_missing_values)
```

```python
train.loc[train[columns_to_impute].isnull().any(axis=1), columns_to_impute] = predicted_values[columns_to_impute]
```
